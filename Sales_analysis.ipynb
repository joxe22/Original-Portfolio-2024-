### Z-Score
Un Z-Score, tambin conocido como puntuacin Z, es una medida estadstica que indica cuntas desviaciones estndar un punto de datos especfico est por encima o por debajo de la media del conjunto de datos.

En este sentido, podemos calcular la puntuacin Z de cada dato usando la funcin zscore de scipy y compararla con un umbral para determinar qu valores son considerados atpicos. Por lo general se establece un umbral de 3, por lo que aquellos puntos de datos cuya puntuacin Z absoluta sea superior a 3 son outliers.

```python
#import zscore function

from scipy.stats import zscore

#calculate z-score for each data point and compute its absolute value
z_scores = zscore(df['Profit'])
abs_z_scores = np.abs(z_scores)

#Selecting the outliers using a Threshold of 3
outliers = df[abs_z_scores > 3]
outliers.head()
```

```python
#obtaining number of outliers
print(f'Number of outliers: {len(outliers)}')
```

```python
def find_columns_without_outliers(df):
    columns_without_outliers = []

    for column in df.select_dtypes(include=[float, int]).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

        if outliers.empty:
            columns_without_outliers.append(column)

    return columns_without_outliers

# Encontrar columnas sin outliers
columns_without_outliers = find_columns_without_outliers(df)
print(f'Columnas sin outliers: {columns_without_outliers}')
```

### Copy the dataset to impute values and replace it with outliers
La imputacin implica reemplazar los valores atpicos con otros valores como la mediana o la media. Esto se suele hacer cuando queremos conservar la mayor cantidad de datos, pero eliminando el efecto de los outliers.

```python
#Dropping columns without outliers
df_filtered = df.drop(['Manufacturing Price', 'Sale Price', 'Month Number'],axis=1)
```

```python
df_filtered.head()
```

```python
# Droping Categorical columns from df_filtered
df_filtered.drop(['Segment', 'Country', 'Product', 'Month Name'], axis=1)
```

```python
df_filtered.head()
```

```python
# impute and calculate de median
median_value = df_filtered[['Units Sold', 'Gross Sales', 'COGS', 'Profit']].median()
```

```python
# Impute outliers with the median

for column in ['Units Sold', 'Gross Sales', 'COGS', 'Profit']:
    df_filtered.loc[outliers.index, column] = median_value[column]
```

### Winzorizing

Winzorizar  La winsorizacin es una tcnica que reemplaza los valores atpicos por el valor ms cercano que no se considera un outlier segn ciertos criterios.

Por ejemplo, podemos reemplazar los valores atpicos por el valor del percentil 95 o 5 ms cercano, reduciendo el efecto de los outliers sin eliminarlos por completo.

```python
from scipy.stats.mstats import winsorize
# Copy of our dataframe
df_filtered_win = df_filtered.copy()
#List of the columns to winsorizing
win_columns = ['Units Sold', 'Gross Sales', 'COGS', 'Profit']
# Appliying the columns to winsorizing
for col in win_columns:
    df_filtered_win[col] = winsorize(df_filtered_win[col], limits=[0.05, 0.05])
    
print("Original DataFrame:")
print(df_filtered)

print("\nDataFrame Winsorized:")
print(df_filtered_win)
```
